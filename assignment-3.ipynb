```python
Q2. Write a python program to print the sum and product of the first 10 natural numbers using for and while loop. 
Ans:-
```

```python



sum = 0
product = 1

for i in range(1, 11):
    sum += i
    product *= i

print("The sum of the first 10 natural numbers is:", sum)
print("The product of the first 10 natural numbers is:", product)
```

```python

sum = 0
product = 1
i = 1

while i <= 10:
    sum += i
    product *= i
    i += 1

print("The sum of the first 10 natural numbers is:", sum)
print("The product of the first 10 natural numbers is:", product)
```

```python
Q3. Create a python program to compute the electricity bill for a household. The per-unit charges in rupees are as follows: For the first 100 units, the user will be charged Rs. 4.5 per unit, for the next 100 units, the user will be charged Rs. 6 per unit, and for the next 100 units, the user will be charged Rs. 10 per unit, After 300 units and above the user will be charged Rs. 20 per unit.   
You are required to take the units of electricity consumed in a month from the user as input.   
Your program must pass this test case: when the unit of electricity consumed by the user in a month is 310, the total electricity bill should be 2250. 
Ans:-
```

```python

def calculate_electricity_bill(units):
    bill = 0
    
    if units <= 100:
        bill = units * 4.5
    elif units <= 200:
        bill = (100 * 4.5) + (units - 100) * 6
    elif units <= 300:
        bill = (100 * 4.5) + (100 * 6) + (units - 200) * 10
    else:
        bill = (100 * 4.5) + (100 * 6) + (100 * 10) + (units - 300) * 20
    
    return bill


units_consumed = int(input("Enter the number of units consumed: "))


total_bill = calculate_electricity_bill(units_consumed)


print(f"The total electricity bill for {units_consumed} units is: Rs. {total_bill:.2f}")

```

```python
Q4. Create a list of numbers from 1 to 100. Use for loop and while loop to calculate the cube of each number and if the cube of that number is divisible by 4 or 5 then append that number in a list and print that list. 
```

```python

numbers = []
cube_divisible = []

for i in range(1, 101):
    numbers.append(i)
    cube = i ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        cube_divisible.append(cube)

print("The list of cubes that are divisible by 4 or 5:", cube_divisible)
```

```python
#here is same for while loop
numbers = []
cube_divisible = []
i = 1

while i <= 100:
    numbers.append(i)
    cube = i ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        cube_divisible.append(cube)
    i += 1

print("The list of cubes that are divisible by 4 or 5:", cube_divisible)
```

```python
Q5. Write a program to filter count vowels in the below-given string. string = "I want to become a data scientist"
Ans:
```

```python

string = "I want to become a data scientist"
vowels = "aeiouAEIOU"
count = 0

for char in string:
    if char in vowels:
        count += 1

print("The number of vowels in the string is:", count)
```
